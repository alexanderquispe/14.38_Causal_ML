# Simple Exercise on Overfitting


First set p=n

```r

set.seed(123)
n = 1000

p = n
X<- matrix(rnorm(n*p), n, p)
Y<- rnorm(n)

print("p/n is")
print(p/n)
print("R2 is")
print(summary(lm(Y~X))$r.squared)
print("Adjusted R2 is")
print(summary(lm(Y~X))$adj.r.squared)
```

```r
print(summary(lm(Y~X)))
```
